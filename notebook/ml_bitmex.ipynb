# Analyze Market Depth


### Load Data

```python
import os
import csv
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt

os.chdir("C:/Users/research/src/ml-bitmex/log")
print('Read data from', os.getcwd())
market_df = pd.read_csv('Market_Dept.csv')
```

### Extract and Group Sell and Buy data

```python
def group_price(market_df,op):
    market_side =market_df[market_df["side"] == op] 
    return market_side.sort_values(["price"]).groupby(['price']).sum()

def plot_order_book(df):
    plt.bar(df.index, df['size'])
    plt.show()
```

```python
df_sell = group_price(market_df, "Sell")
df_buy = group_price(market_df, "Buy")
plot_order_book(df_sell.head(10))
plot_order_book(df_buy.tail(10))
```
